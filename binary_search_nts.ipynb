## Binary search is a way to take a big problem and cut it in half<br>
## instead of searhing a whole list ,we half the list based on the condition ,like finding a target in a sorted list<br>
## It works by repeatedly dviding the search interval in half 

# which gives us <b>O(log n) time complexity</b>

```python
x=[i for i in range(100)]
y=np.log10(x)
plt.figure(figsize=(28,3))

plt.xlabel("number of input values")
plt.ylabel("number of operations")
plt.title("O(log n)")

plt.plot(x,y)
```

We do this by getting a mid value and starting there .This is done by getting the sum of of the first and last values in the array and divding that by 2<br>
we then check if that value is less or greater than our target ,if less ,drop everything from the left ,if greater,drop everything from the right and repeat   

```python
nums=[-1,0,3,5,9,12]

def binary_search(nums:list,target:int):
    left,right=0,len(nums) -1 #intital indicies for searching ;right is the last index
    while left <= right:
        mid=(left+right)//2 
        if nums[mid] > target: # if the middle element is greater than the target
            right=mid-1  #narrow the search to the left half by adjusting the right index
        elif nums[mid] < target:
            left=mid+1 #if the middle element is less than the target ,narrow the search to the right half by adjusting the left index 
        else:
            return mid #else return mid , that means the value is neither  less or greater than the target ,so it must be equal 
    return -1 #return -1 if target isnt found 
```

```python
binary_search(nums,6)#the number 6 doesnt exist in this array
```

```python
binary_search(nums,9)#the number 9 can be found at index point 4
```
